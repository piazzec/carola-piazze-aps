## Desarrollo
Se reemplazan los elementos del circuito RLC por impedancias Z para poder trabajar. EL circuito está compuesto por dos secciones, cada una con tres impedancias en paralelo. Se resuelve cada sección por separado y se halla $Z_{p1}$ y $Z_{p2}$, que luego se utilizan en un divisor de tensiones para hallar la función de transferencia **T(s)**

##### 1. Impedancias elementales (en función de $s$)  


$$
\begin{aligned}
Z_{1} &= Z_{C a} = \frac{1}{s\,C\,a} \\[6pt]
Z_{2} &= Z_{L b} = \frac{sL}{b} \\[6pt]
Z_{3} &= Z_{R d} = \frac{R}{d} \\[6pt]
Z_{4} &= Z_{R (1-d)} = \frac{R}{1-d} \\[6pt]
Z_{5} &= Z_{L (1-b)} = \frac{sL}{1-b} \\[6pt]
Z_{6} &= Z_{C (1-a)} = \frac{1}{s\,C\,(1-a)}
\end{aligned}
$$



##### 2. Impedancias de los dos bloques en paralelo

**Paralelo izquierdo (los tres elementos etiquetados 1,2,3)**
La impedancia equivalente $Z_{p1}$ es:
$$
Z_{p1} = \left(\frac{1}{Z_1}+\frac{1}{Z_2}+\frac{1}{Z_3}\right)^{-1}
$$
Sustituyendo $Z_1,Z_2,Z_3$:
$$
Z_{p1} = \left( s C a + \frac{b}{sL} + \frac{d}{R} \right)^{-1}
$$  
lo  cual es equivalente a:
$$Z_{p1} = \dfrac{sR L}{s^2 C R L a + R b + s L d}$$

que se obtiene multiplicando numerador y denominador por $sR L$.



**Paralelo derecho (los Z 4,5 y 6)**
La impedancia equivalente $Z_{p2}$ es:
$$
Z_{p2} = \left(\frac{1}{Z_4}+\frac{1}{Z_5}+\frac{1}{Z_6}\right)^{-1}
$$
Sustituyendo $Z_4,Z_5,Z_6$:
$$
Z_{p2} = \left( \frac{1-d}{R} + \frac{1-b}{sL} + sC(1-a) \right)^{-1}
$$  
que es igual a  
$$Z_{p2} = \dfrac{sR L}{(1-d)s^2 L C + (1-b)R + s(1-a) R L C}$$ 



##### 3. Divisor de tensiones — $T(s)=\dfrac{V_2}{V_1}$

El circuito entre $V_1$ y $V_2$ es un divisor simple entre las dos impedancias en serie $Z_{p1}$ y $Z_{p2}$. Entonces:

$$
T(s)=\frac{V_2}{V_1} = \frac{Z_{p2}}{Z_{p1}+Z_{p2}}
$$


Entonces: 

**Primer paralelo:**

$$
Z_{p1} = \frac{s R L}{s^2 C R L a + R b + s L d} = \frac{s R L}{A(s)}
$$

donde  
$$A(s) = s^2 C R L a + R b + s L d$$


**Segundo paralelo:**

$$
Z_{p2} = \frac{s R L}{(1-d)s^2 L C + (1-b)R + s(1-a) R L C} = \frac{s R L}{B(s)}
$$

donde  
$$B(s) = (1-d)s^2 L C + (1-b)R + s(1-a)R L C$$


##### 4. Divisor de tensiones

Para el divisor de tensiones:

$$
\frac{V_2}{V_1} = \frac{Z_2}{Z_1 + Z_2} = \frac{Z_{p2}}{Z_{p1} + Z_{p2}}
$$

Sustituyendo $Z_{p1}$ y $Z_{p2}$ en términos de $A(s)$ y $B(s)$:

$$
T(s) = \frac{\dfrac{s R L}{B(s)}}{\dfrac{s R L}{A(s)} + \dfrac{s R L}{B(s)}} = \frac{A(s)}{A(s)+B(s)}
$$


##### 5. Desarrollo del numerador y denominador

Expresando los términos:

$$
A(s) = s^2 C R L a + R b + s L d
$$

$$
B(s) = (1-d)s^2 L C + (1-b)R + s(1-a)R L C
$$

Sumando ambos:

$$
A(s) + B(s) = s^2 C R L a + s^2 L C (1-d) + R b + R (1-b) + s L d + s(1-a) R L C
$$

Luego:

$$
A(s) + B(s) = s^2 R L C + s L + R
$$


#### Finalmente:

Reemplazando en la expresión de $T(s)$:

$$
T(s) = \frac{s^2 R L C a + s L d + R b}{s^2 R L C + s L + R}
$$

Dividiendo numerador y denominador por $R L C$ para dejar normalizado el término $s^2$:

$$
T(s) = \frac{s^2 a + \dfrac{s d}{R C} + \dfrac{b}{L C}}{s^2 + \dfrac{s}{R C} + \dfrac{1}{L C}}
$$


### Normalizar y escribir en forma bicuadrática (SOS)

La forma estándar de una sección bicuadrática (second-order section) es:

$$
T(s)=k\cdot \frac{s^2 + s\frac{\omega_{0Z}}{Q_Z} + \omega_{0Z}^2}{s^2 + s\frac{\omega_{0P}}{Q_P} + \omega_{0P}^2}
$$

Para llevar una fracción racional a esa forma, se factoriza el coeficiente del término cuadrático y se comparan coeficientes, identificando $k$, $\omega_{0Z}$, $Q_Z$, $\omega_{0P}$ y $Q_P$ según corresponda.


### Cálculo de ceros y polos
 Despejamos los coeficientes de la T(s) que encontramos usando la forma estándar de la SOS:
##### Ceros

$$
\frac{\omega_{0Z}}{Q_Z} = \frac{d}{R_C C_A}
$$

$$
\omega_{0Z} = \sqrt{\frac{b}{L_C A}}
$$

$$
\Rightarrow Q_Z = \frac{\omega_{0Z} R_C C_A}{d}
$$

$$
\omega_{0Z} = \frac{R_C C_A}{d} \sqrt{\frac{b}{L_C A}}
$$

$$
Q_Z = \frac{R_C C_A}{d} \sqrt{\frac{b}{L_C A}}
$$


##### Polos

$$
\omega_{0P} = \frac{1}{\sqrt{L C}}
$$

$$
\frac{\omega_{0P}}{Q_P} = \frac{1}{R C}
$$

$$
\Rightarrow Q_P = \omega_{0P} (R C) = \frac{R C}{\sqrt{L C}}
$$




```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

#funciones de transferencia, me armo un diccionario con las transferencias, 
#pongo los coeficientes que corresponden
sistemas = {
"T₁(s) = (s² + 9)/(s² + √2·s + 1)": ([1, 0, 9], [1, np.sqrt(2), 1]),
"T₂(s) = (s² + 1/9)/(s² + 0.2·s + 1)": ([1, 0, 1/9], [1, 0.2, 1]),
"T₃(s) = (s² + 0.2·s + 1)/(s² + √2·s + 1)": ([1, 0.2, 1], [1, np.sqrt(2), 1]),
}
for T, (b, a) in sistemas.items():
    
    #saco los polos y ceros
    z, p, k = signal.tf2zpk(b, a)
    
    #diagrama de polos y ceros
    plt.figure(figsize=(12, 4))
    plt.plot(np.real(p), np.imag(p), 'x', markersize=10, label='Polos')
    
    #z es el arreglo de ceros como algunas funciones de transferencia pueden no tener ceros
    #hay que chequear si hay antes de plottear
    if len(z) > 0:
        plt.plot(np.real(z), np.imag(z), 'o', markersize=10, fillstyle='none', label='Ceros')
    plt.title(f'{T} - Diagrama de Polos y Ceros')
    plt.xlabel('Parte Real (σ)', fontsize=11)
    plt.ylabel('Parte Imaginaria (jω)', fontsize=11)
    plt.axhline(0, color='gray', lw=1)
    plt.axvline(0, color='gray', lw=1)
    plt.grid(True, ls='--', color='lightgray')
    plt.legend(frameon=False)

    #Rta en frecuencia
    w, h = signal.freqs(b=b, a=a, worN=np.logspace(-1, 2, 1000))
    fase = np.unwrap(np.angle(h))
    #corrige esos saltos, sumando o restando múltiplos de 2π para que la fase quede continua y suave
    gd = -np.diff(fase) / np.diff(w)
    #calcula el retardo de grupo
    
    #magnitud
    plt.figure(figsize=(12, 4))
    plt.semilogx(w, 20 * np.log10(abs(h)), color='hotpink', linewidth=2)
    #el eje X se comprime logarítmicamente, mostrando mejor cómo cambia 
    #la fase en todo el rango de frecuencias.
    plt.title(f'{T} - Respuesta en Magnitud')
    plt.xlabel('Pulsación angular [rad/s]')
    plt.ylabel('|H(jω)| [dB]')
    plt.grid(True, which='both', ls='--')
    
    #fase
    plt.figure(figsize=(12, 4))
    plt.semilogx(w, np.degrees(fase), color='hotpink', linewidth=2)
    plt.title(f'{T} - Fase')
    plt.xlabel('Pulsación angular [rad/s]')
    plt.ylabel('Fase [°]')
    plt.grid(True, which='both', ls='--')
    
    plt.tight_layout()
    plt.show()
```
